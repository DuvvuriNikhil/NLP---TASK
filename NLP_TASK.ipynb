```python
text = """In the modern world, communication is everywhere—emails, messages, calls, and even quick reactions online! But have you ever stopped to think: are we really communicating effectively? Sometimes, a simple “Okay.” can sound polite… or rude, depending on context. Isn’t that strange? Language is powerful; it shapes emotions, decisions, and relationships. For example, consider this: “Let’s eat, grandma!” versus “Let’s eat grandma!” — a comma literally saves a life. So, punctuation isn’t just grammar; it’s meaning, tone, and clarity all combined.
Life, however, isn’t only about rules—it’s also about expression. You can write formally; you can write casually; you can even mix styles—why not? Questions arise: What matters more—accuracy or creativity? The answer is: both! A well-written sentence can inform, persuade, or even inspire… sometimes all at once. And yes, mistakes happen; everyone makes them! But that’s okay—learning comes from trying, failing, and improving. So, keep writing, keep experimenting, and most importantly—enjoy the process"""
```

**Punctuation Removal**

```python
cleaned_text = "".join(char for char in text if char.isalnum() or char == " ")
print(cleaned_text)
```

**Tokenization**

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(cleaned_text)
print(tokens)
```

**Stop Word Removal**

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
print(filtered_tokens)
```

**Stemming**

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stem_words = [stemmer.stem(token) for token in filtered_tokens]
print(stem_words)
```

**Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemma_words = [lemmatizer.lemmatize(token, pos='v') for token in filtered_tokens]
print(lemma_words)
```

```python
converted_text = " ".join(lemma_words)
print(converted_text)
```

**Bag of Words**

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [converted_text]
vectorizer = CountVectorizer()
vectorized_text = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names_out())
print(vectorized_text.toarray())
```

**TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus1 = [converted_text]
vectorizer1 = TfidfVectorizer()
vectorized_text1 = vectorizer1.fit_transform(corpus1)
print(vectorizer1.get_feature_names_out())
print(vectorized_text1.toarray())
```

**Word 2 Vec**

```python
#Large Text
text1 = """In today’s rapidly evolving digital era, the importance of continuous learning cannot be overstated. Technology is advancing at an unprecedented pace, transforming the way we live, work, and communicate. As a result, individuals must actively seek opportunities to upgrade their skills and expand their knowledge. Whether it is learning a new programming language, improving communication abilities, or understanding emerging trends like artificial intelligence and data science, staying updated is essential for both personal and professional growth. Those who embrace learning as a lifelong habit are better equipped to adapt to changes and overcome challenges in a competitive environment.
Moreover, consistency plays a crucial role in achieving long-term success. It is not the intensity of effort that matters most, but the regularity with which one practices and improves. Small, consistent actions performed daily can lead to significant results over time. For instance, dedicating even an hour each day to learning or practicing a skill can result in noticeable improvement within a few months. Alongside consistency, maintaining discipline and focus helps individuals avoid distractions and stay aligned with their goals. Setting clear objectives and tracking progress can further enhance motivation and accountability.
Equally important is maintaining a healthy balance between work and personal well-being. In the pursuit of success, people often neglect their physical and mental health, which can lead to burnout and reduced productivity. Engaging in activities such as exercise, meditation, or hobbies can help refresh the mind and improve overall efficiency. A balanced lifestyle ensures that individuals remain energetic, creative, and resilient in the face of challenges."""
```

```python
#Removing punctuations
clean_text = "".join(char for char in text1 if char.isalnum() or char == " ")
print(clean_text)
```

```python
#Sentence Tokenize
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(clean_text.lower())
print(sentences)
```

```python
#Word Tokenize
from nltk.tokenize import word_tokenize
words = [word_tokenize(sentence) for sentence in sentences]
print(words)
```

```python
from gensim.models import Word2Vec
w2v_model = Word2Vec(sentences=words,
               vector_size=100,
               window=5,
               min_count=1,
               epochs=100)
```

**Checking Word Similarity**

```python
print("Similarity between the words importance and continuous" ,w2v_model.wv.similarity("importance", "continuous"))
```

```python
print("Similarity between the words overstated and technology" ,w2v_model.wv.similarity("overstated", "technology"))
```

```python
print("Similarity between the words lerning and advancing" ,w2v_model.wv.similarity("learning", "advancing"))
```

```python
print("Similarity between the words digital and unprecedented" ,w2v_model.wv.similarity("digital", "unprecedented"))
```
